## 17. Male ratio per occupation sorted most to least


```python
def gender_code(g):
    return 1 if(g == 'M') else 0

users['gender_code'] = users.gender.apply(gender_code)

users.groupby(by=['occupation']).gender_code.sum() / users.occupation.value_counts() * 100
```

## 18. Minimum and Max Age per occupation

```python
o1  = users.groupby('occupation').agg({'age': min})
o1.columns = ['min_age']
o2 = users.groupby('occupation').agg({'age': max})
o2.columns = ['max_age']
o1.join(o2)
```

## 19. Mean age by occupation and gender

```python
m  = users.groupby(['occupation', 'gender'])['age'].mean()
pd.DataFrame(m)
```

## 20. Percentage of Men and Women by Occupation

```python
p = pd.DataFrame((users.groupby(['occupation', 'gender'])['age'].count() / users.groupby(['occupation'])['age'].count() * 100))
p.columns = ['age %']
p
```
