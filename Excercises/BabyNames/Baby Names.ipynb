```python
import pandas as pd
```

```python
bnames = pd.read_csv('babynames.csv', sep=',')
bnames.head()
len(bnames)
```

```python
bnames.head(10)
```

```python
bnames.drop(['Unnamed: 0'], axis = 1, inplace=True)
bnames.head()
```

```python
malenames =  len(bnames[bnames['Gender'] == "M"].groupby('Name'))
femalenames = len(bnames[bnames['Gender'] == "F"].groupby('Name'))
print('Male Name: ', malenames)
print('Female Name: ', femalenames)


```

```python
#Group By Names
names = bnames.groupby('Name').head(1)
names
```

```python
# Delete column Year
# Why? - Becasue it is irrelevant now once groupby is done for the Name
names.drop(['Year'], axis = 1, inplace=True)
```

```python
# first 5 rows
names.head()
```

```python
print('Size:', names.size) #not sure what size is?
print('Dimensions:', names.shape)
```

```python
names.sort_values(by = 'Count', ascending = False)
names.head()
```
